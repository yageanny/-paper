```python
import pandas as pd
import pandas as pd

# 读取Stata的dta文件
df = pd.read_excel('D:/2023年寒假/司法大数据/原文.xlsx')

df.head()
```

```python
###########分词方案一：

import pandas as pd
import jieba


stop_list = []
with open('停用词表.txt', 'r', encoding='utf-8') as f:
    for line in f.readlines():
         stop_list.append(line.replace('\n', ''))

cut = open('cut.csv', 'w', encoding='utf-8')

for idx, row in df.iterrows():
    words= jieba.cut(str(row['原文']))

    for word in words:
        if word not in stop_list:
            cut.write(word + ' ')
        else:
            print(word)
            
    cut.write('\n')

cut.close()
df['judge_cut']=''
```
