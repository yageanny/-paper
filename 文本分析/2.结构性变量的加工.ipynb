```python
import pandas as pd

df = pd.read_csv('明星信息1.csv', encoding='utf-8')
df.head()
```

```python
df.head()
```

```python
################一、通过有监督学习构建出因变量（自己知道得到的是什么）
import numpy as np

######生成年龄
df['birthyear']=df['birthday'].str.split('年')
df['birthyear1']=df['birthyear'].apply(lambda x:x[0] if x is not np.nan else '')
df['birthyear2']=pd.to_numeric(df['birthyear1'],errors='coerce')
df['age']=2020 - df['birthyear2']
df.drop(['birthyear','birthyear1','birthyear2'], axis=1, inplace=True)
#######生成代表作数量和荣誉数量
df['work_amount']=df['main_works'].str.count('\n')+1
df['achieve_amount']=df['main_achieve'].str.count('\n')+1

#######生成是否有nickname
df['is_nickname']=df['nickname'].apply(lambda x:0 if x is np.nan else 1)

#######生产occupation类别的多样性
df['occupation_amount']=df['occupation'].str.count(' ')+1

#######生成是否有慈善行为
df['achieve_charity']=df['main_achieve'].apply(lambda x:1 if '慈善' in str(x) else 0)
df['judge_charity']=df['judge'].apply(lambda x:1 if '慈善' in str(x) else 0)

df['charity'] = df.apply(lambda x:0 if (x['achieve_charity']==0) and (x['judge_charity']==0) else 1, axis=1)

#######生成获奖是否与香港相关
df['achieve_hongkong']=df['main_achieve'].apply(lambda x:1 if '香港' in str(x) else 0)

#######生成是否有春晚经历
df['achieve_chunwan']=df['main_achieve'].apply(lambda x:1 if( '春晚' or '春节联欢晚会') in str(x) else 0)

df.head()
# df.to_csv('明星信息2.csv')
```

## 研究明星慈善行为与声望的关系

```python
charity =['慈善','爱心','捐赠','捐款','捐钱','捐助']
#var_name=['if_charity1','if_charity2','if_charity3','if_charity4','if_charity5','if_charity6']
for (i, w) in enumerate(charity):
    df['if_charity{}'.format(i)]=df['judge'].apply(lambda x:1 if str(w) in str(x) else 0)
```

```python
charity =['慈善','爱心','捐赠','捐款','捐钱','捐助']
#var_name=['if_charity1','if_charity2','if_charity3','if_charity4','if_charity5','if_charity6']
for (i, w) in enumerate(charity):
    df['num_charity{}'.format(i)]=df['judge'].apply(lambda x: str(x).count(str(w)) )
```

```python
respect =['榜样','模范','典范','楷模','表率','典型','声誉','名望','名气','声名','名誉','荣誉']
var_name=['if_by','if_mf','if_df','if_km','if_bs','if_dx','if_sy','if_mw','if_mq','if_sm','if_my','if_ry']
for (w, n) in zip(respect,var_name):
    df['{}'.format(n)]=df['judge'].apply(lambda x:1 if str(w) in str(x) else 0)
```
