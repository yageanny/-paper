### 命名实体识别
命名实体识别（Named Entity Recognition，简称NER），又称作“专名识别”，是指识别文本中具有特定意义的实体，主要包括人名、地名、机构名、专有名词等。简单的讲，就是识别自然文本中的实体指称的边界和类别。


开源 NER (Named-entity recognition)工具包有：

StanfordCoreNLP

 https://stanfordnlp.github.io/CoreNLP/

Hanlp

http://www.hanlp.com/

foolNLTK

https://github.com/rockyzhengwu/FoolNLTK

LTP
http://www.ltp-cloud.com/


```python
import pandas as pd

df = pd.read_csv('明星信息2.csv', encoding='utf-8')
df.head()
```

```python
#########命名实体抽取（地名、日期）FoolNLTK


#######需提前安装TensorFlow。（TensorFlow安装提醒:TensorFlow1.3版本以上，2.0（不含）以下，
###因为需要tensorflow.contrib模块，所以对应的Python版本需要为64位，而不是32位，并且为3.8（不含）以下，
###因为Python3.8以上对应为TensorFlow为2.0以上的。）

import fool

res1 = []
site1 = []
time1 = []
for i in df['judge_cut']:
    res = fool.analysis(str(i))
    res1.append(res)
    site = str(res).count('location')
    site1.append(site)
    time = str(res).count('time')
    time1.append(time)
    

df['judge_res']=res1
df['judge_site_n']=site1
df['judge_time_n']=time1
print(res1)
```
